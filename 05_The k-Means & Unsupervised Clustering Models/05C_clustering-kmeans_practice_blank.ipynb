# 05 | Cluster Analysis with k-Means

## Load the Data

- Simply execute the following lines of code to load the data
- This dataset contains **statistics** (columns)
- About **Car Models** (rows)

```python
import seaborn as sns #!

df_mpg = sns.load_dataset(name='mpg', index_col='name')
df_mpg.sample(5)
```

## Feature Selection

- Select 2 columns for the modelling

```python
features = df_mpg[['mpg' , 'acceleration']]
```

```python
features
```

## Data `preprocessing`

- Do you need to *transform* the data
- To get a **truthful insight** of the model?

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(features)
```

```python
X_scaler =scaler.transform(features)
X_scaler
```

## `KMeans()` Model in Python

### Fit the Model

- To cluster 3 groups

```python
from sklearn.cluster import KMeans
```

```python
kmean =KMeans(n_clusters= 3)
```

```python
kmean.fit(X=features)
```

### Calculate Predictions

```python
kmean.predict(X=features)
```

### Create a New DataFrame for the Predictions

```python
df_predict = features.copy()
df_predict['pred_features'] = kmean.predict(X=features)
```

```python
df_predict
```

### Visualize the Model

- You colour the points with the the value of the predictions

```python
sns.scatterplot(x ='mpg' , y= 'acceleration', hue= 'pred_features' , palette= 'Set1',data= df_predict )
```

## Other `Clustering` Models in Python

- Visit the `sklearn` website [here ↗](https://scikit-learn.org/stable/)
- **Pick 2 new models** and compute the *Clustering*

### Other Model 1

#### Fit the Model

- To cluster 3 groups

```python
from sklearn.cluster import SpectralClustering
```

```python
model_sp = SpectralClustering(n_clusters=3)
```

```python
model_sp.fit(X=X_scaler)
```

#### Calculate Predictions

```python
model_sp.fit_predict(X_scaler)
```

#### Create a New Column for the Predictions

```python
df_predict['pred_sp'] =model_sp.fit_predict(X_scaler)
```

```python
df_predict
```

#### Visualize the Model

```python
sns.scatterplot(x ='mpg' , y= 'acceleration', hue= 'pred_sp' , palette= 'Set1',data= df_predict )
```

### Other Model 2

#### Fit the Model

- To cluster 3 groups

```python
from sklearn.cluster import MeanShift
```

```python
model_shift = MeanShift(cluster_all=3)
```

#### Calculate Predictions

```python
model_shift.fit_predict(X_scaler)
```

#### Create a New Column for the Predictions

```python
df_predict['pred_shift'] = model_shift.fit_predict(X_scaler)
```

```python
df_predict
```

```python
sns.scatterplot(x ='mpg' , y= 'acceleration', hue= 'pred_shift' , palette= 'Set1',data= df_predict )
```

#### Visualize the Model

```python
import matplotlib.pyplot as plt
```

```python
fig , (ax1, ax2 ,ax3) = plt.subplots(1 ,3 ,figsize=(20 ,7))

ax1 .set_title('kmean')
sns.scatterplot(x ='mpg' , y= 'acceleration', hue= 'pred_features' ,data= df_predict , palette ='Set1' , ax=ax1)


ax2 .set_title(' spectral')
sns.scatterplot(x ='mpg' , y= 'acceleration', hue= 'pred_sp' ,data= df_predict, palette= 'Set1', ax=ax2 )



ax3.set_title('shfit')
sns.scatterplot(x ='mpg' , y= 'acceleration', hue= 'pred_shift',data= df_predict ,palette='Set1' , ax=ax3 )
```
